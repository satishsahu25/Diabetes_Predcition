```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv(r"heart.csv")
dataset.head()
```

```python
dataset.describe()
```

```python
catfeature=[]
numfeature=[]

for column in dataset.columns:
    if dataset[column].nunique()<=10:
        catfeature.append(column)
    else:
        numfeature.append(column)
```

```python
catfeature.remove('HeartDisease')
```

```python
newdata=pd.get_dummies(dataset,columns=catfeature,drop_first=True)
newdata
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
newdata[numfeature]=ss.fit_transform(newdata[numfeature])
```

```python
newdata
```

# Heart Disease vs Normal 

```python
plt.figure(figsize=(5,5))
sns.countplot(dataset['HeartDisease'])
```

# Correlation Matrix

```python
plt.figure(figsize=(17,10))
sns.heatmap(newdata.corr(),annot=True)
```

# Heart Disease as per Sex

```python
plt.figure(figsize=(5,5))
sns.countplot(newdata['HeartDisease'])
plt.xticks([0,1],['Female','Male'])
```

```python
sns.countplot(dataset['Sex'])
#labelling of sex
plt.xticks([0,1],['Female','Male'])
plt.show()
```

# Gender Distribution

```python
sns.countplot(x="Sex",hue="HeartDisease",data=dataset)
plt.xticks([1,0],['Male','Female'])
plt.legend(labels=['Normal','HeartDisease'])
plt.show()
```

# Age Distribution

```python
sns.distplot(dataset['Age'],bins=20)
plt.show()
```

# Chest Pain Distribution

```python
sns.countplot(dataset['ChestPainType'])
plt.xticks([0,1,2,3],['TA','ATA','NAP','ASY'])
plt.show()
```

# ChestPain Distribution as per Heart Disease

```python
sns.countplot(x="ChestPainType",hue="HeartDisease",data=dataset)
plt.xticks([0,1,2,3],['TA','ATA','NAP','ASY'])
plt.legend(labels=['Normal','HeartDisease'])
plt.show()
```

```python
sns.countplot(x="FastingBS",hue="HeartDisease",data=dataset)
plt.xticks([0,1],['High','Low'])
plt.legend(labels=['Normal','HeartDisease'])
plt.show()
```
